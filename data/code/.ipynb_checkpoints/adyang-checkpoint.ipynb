# Data Visualization of 1-Minute Time Series for BTC

```python
# Imports
import pandas as pd
import numpy as np
import random
from datetime import datetime
from matplotlib import pyplot

random.seed(0)
```

```python
# Loading one data file

data_file = '../data/gemini_BTCUSD_2021_1min.csv'
data_2021 = pd.read_csv(data_file, skiprows=[0], low_memory=False)
print(data_2021)
```

```python
# Aggregating data

years = range(2015, 2021)

frames = []

for year in years:
    data_file = "../data/gemini_BTCUSD_%s_1min.csv" % str(year)
    frames.append(pd.read_csv(data_file, skiprows=[0], low_memory=False))

frames.append(data_2021)

raw_data = pd.concat(frames)

print(raw_data)
```

```python
# Dropping columns and a row

# raw_data = raw_data.drop(labels=['High', 'Low', 'Close', 'Volume', 'Symbol', 'Unix Timestamp'], axis=1)
raw_data['Date'] = raw_data['Date'].apply(lambda dateString: datetime.strptime(dateString, "%Y-%m-%d %H:%M:%S"))
print(raw_data)
```

```python
# Plotting

raw_data.plot(x='Date')
pyplot.show()
```

```python
# Splitting into k-minute intervals and checking 10 samples

SAMPLE_SIZE = 10

def showcaseIntervalSize(intervalSize):
    intervals = np.array_split(raw_data, raw_data.size // intervalSize)
    L = len(intervals)
    for i in range(SAMPLE_SIZE):
        intervals[random.randint(0, L)].plot(x='Date')
        pyplot.show()
```

```python
# Cross-validation 10/90 training testing split
R_STATE = 42 # (deterministic id for reproducible results)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(raw_data.drop(labels=['Open'], axis=1), raw_data.drop(labels=['Date'], axis=1), test_size=0.1, random_state = R_STATE)
```

```python
# theoretical max function with no overlapping trades

def theoretical_max_gain_no_overlap(data,  trade_factor = 1, price_type = 'Open'):
    labels = list(filter(lambda label: label != price_type, ['High', 'Low', 'Close', 'Volume']))
#     filter_data = data[(data['Date'] >= start) & (data['Date'] <= end)]
    print(data)
    diff = data.drop(columns=labels, axis=1).diff()
    diff = diff.drop(index=diff.index[0])
    diff = diff[diff[price_type] > 0.0]
    return diff.sum() * trade_factor

# print(theoretical_max_gain_no_overlap(raw_data, datetime(year=2020, month=1, day=1), datetime(year=2020, month=12, day=31), 10))
```

```python
def theoretical_max_gain(data, start, end, stocks_per_buy = 1, price_type = 'Open'):
    labels = list(filter(lambda label: label != price_type, [ 'High', 'Low', 'Close', 'Volume']))
#     filter_data = data[(data['Date'] >= start) & (data['Date'] <= end)].drop(columns=labels, axis=1)
    print(data)
    max_data = data[::-1].cummax()
    combined_data = pd.concat([data, max_data], ignore_index=True, axis=1)
    diff = combined_data.apply(lambda row: 0.0 if row[1] - row[0] <= 0.0 else row[1] - row[0], axis=1)
    return diff.sum() * stocks_per_buy

# print(raw_data)
# print(theoretical_max_gain(raw_data, datetime(year=2020, month=1, day=1), datetime(year=2020, month=12, day=31), 10))
```
